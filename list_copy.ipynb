```python
import arcpy
import os
# works on current workspace 
ws = r"R:\GEOG491_16440_FALL2025\Student_Data\gshaffer\Week2\Lab2\Ex06"
fgdb = "Copy.gdb"
arcpy.CreateFileGDB_management(ws, fgdb)
arcpy.env.workspace = ws 
```

```python
# getting a list of the feature classes
fclist = arcpy.ListFeatureClasses()
```

```python
# getting the geometry of each feature class and the spatial reference name for each feature class
for fc in fclist:
    fcname= arcpy.da.Describe(fc)["baseName"]          #property of the feature class 
    newfc = os.path.join(ws, fgdb, fcname)             #creating the correct path
    arcpy.CopyFeatures_management(fc, newfc)
```
