```python
# allows use of arcpy toolkit 
import arcpy
# setting correct workspace in R drive under exercise 8 
arcpy.env.workspace = r"R:\GEOG491_16440_FALL2025\Student_Data\gshaffer\Lab3\Ex08"

# the feature class we are using for this code 
fc = "airports.shp"
```

```python
# this is the coding equivalent of Select by Features (or dialog driven queries), .AddFieldDelimiters
delimfield = arcpy.AddFieldDelimiters(fc, "STATE")

# setting up an SQL expression to use to update the data in the STATE section of the attribute table 
sql_exp = delimfield + " <> 'AK'"

# updating the rows in the attribute table to have AK in each STATE box
with arcpy.da.UpdateCursor(fc, ["STATE"], sql_exp) as cursor: 
    for row in cursor: 
        row[0] = "AK"
        cursor.updateRow(row)
```
