```python
import arcpy
import os
# works on current workspace 
# Input workspace (where your data currently is)
input_ws = r"R:\GEOG491_16440_FALL2025\Class_Data\Data2025\PublicDataGeoDatabase.gdb"

# Output workspace (where you want to copy data to)
output_ws = r"R:\GEOG491_16440_FALL2025\Student_Data\gshaffer\Week2\Lab2\Ex06\Results"
arcpy.env.workspace = input_ws

# name for new gdb greated 
fgdb = "Albany_Polygons.gdb"
gdb_path = os.path.join(output_ws, fgdb)

# allowing overwrite in case files already exist 
arcpy.env.overwriteOutput = True
```

```python
# using try-except for error checking 
try: 
    fclist = arcpy.ListFeatureClasses()
    # reading the list and getting information from it 
    
    if not arcpy.Exists(gdb_path):
        arcpy.CreateFileGDB_management(output_ws, fgdb)
        
    for fc in fclist:
        fcdesc = arcpy.da.Describe(fc) 
        fcname = fcdesc["baseName"]
        stype = fcdesc["shapeType"]
            
        # renaming the files (checking if name already includes "_polygon" 
        if stype == "Polygon": 
            if not fcname.lower().endswith("_polygon"):
                new_name = fcname + "_Polygon"
            else: 
                new_name = fcname
        
            # This builds the output path, where the files are going and the new name appended onto the polygon files 
            newfc = os.path.join(gdb_path, new_name)     
    
            # If the output feature class already exists, delete it
            if arcpy.Exists(newfc):
                arcpy.Delete_management(newfc)
    
            #copying the new feature class to the correct file
            arcpy.CopyFeatures_management(fc, newfc)
                
            # Checking for errors by printing the results of the copied files 
            print(f"Copied: {newfc}")  

#  Provide a short explanation of the error or prints another result if the files are not copying correctly 
except arcpy.ExecuteError: 
    print(arcpy.GetMessages(2))
except: 
    print("There has been a nontool error")

```
