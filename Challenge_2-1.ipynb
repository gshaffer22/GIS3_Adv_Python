```python
# importing arcpy & os toolkits 
import arcpy
import os

# Input workspace (where your data currently is)
input_ws = r"R:\GEOG491_16440_FALL2025\Student_Data\gshaffer\Lab4\Ex10"

# Output workspace (where you want to copy data to)
output_ws = r"R:\GEOG491_16440_FALL2025\Student_Data\gshaffer\Lab4\Ex10\Challenge2_Results"
arcpy.env.workspace = input_ws

# name for new gdb greated 
fgdb = "Lab4_Challenge2.gdb"
gdb_path = os.path.join(output_ws, fgdb)
```

```python
# using try-except for error checking 
try: 
    rasterlist = arcpy.ListRasters()
    # reading the list and getting information from it 
    
    if not arcpy.Exists(gdb_path):
        arcpy.CreateFileGDB_management(output_ws, fgdb)
                
    for raster in rasterlist: 
        desc = arcpy.Describe(raster)
        raster_name = desc.baseName        
        new_name = raster_name 
        if not desc.spatialReference or desc.spatialReference.name == "Unknown":
            print(f"Skipping {raster} (no spatial reference)")
            continue

        # Builds output path, where files are going 
        new_rasters = os.path.join(gdb_path, new_name)     

        # If the raster already exists, delete it
        if arcpy.Exists(new_rasters):
            arcpy.Delete_management(new_rasters)
    
        #copying the new feature class to the correct file
        arcpy.CopyRaster_management(raster, new_rasters)
                
        # Checking for errors by printing the results of the copied files 
        print(f"Copied: {new_rasters}")  

#  Provide a short explanation of the error or prints another result if the files are not copying correctly 
except arcpy.ExecuteError: 
    print(arcpy.GetMessages(2))
except: 
    print("There has been a nontool error")
```
