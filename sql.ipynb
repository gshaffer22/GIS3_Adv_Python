```python
# allows use of arcpy toolkit 
import arcpy
# setting correct workspace in R drive under exercise 8 
arcpy.env.workspace = r"R:\GEOG491_16440_FALL2025\Student_Data\gshaffer\Lab3\Ex08"

# the feature class we are using for this code 
fc = "airports.shp"
```

```python
# this is the coding equivalent of Select by Features (or dialog driven queries), .AddFieldDelimiters
delim_field = arcpy.AddFieldDelimiters(fc, "COUNTY")
# establishing the SQL expression that will be used in the for loop 
sql_exp = delim_field + " = 'Anchorage Borough'"
```

```python
# creating a search cursor that will return a list of airports and names for which the SQL expression is true
cursor = arcpy.da.SearchCursor(fc, ["NAME"], sql_exp)
# for loop that will iterate through all of the airports from the feature class
for row in cursor: 
    print(row[0])
```
