```python
# allows use of arcpy toolkit 
import arcpy
# setting correct workspace in R drive under exercise 8 
arcpy.env.workspace = r"R:\GEOG491_16440_FALL2025\Student_Data\gshaffer\Lab3\Challenge"

# feature class for this code: 
fc = "TRANS_PLAN_ADOPTED_2006.shp"

# allowing for overwrite of the code 
arcpy.env.overwriteOutput = True
```

```python
# Add COLLECTOR field to the TRANS_PLAN_ADOPTED_2006 data layer 
# inserting a new record to the attribute table with new name as "COLLECTOR" 
newfield = "COLLECTOR" 
fieldtype = "TEXT" 
fieldname = arcpy.ValidateFieldName(newfield)
arcpy.AddField_management(fc, fieldname, fieldtype)
```

```python
# SQL expression to test for the MAJCOL and MINCOL 
delim_field = arcpy.AddFieldDelimiters(fc, "Layer")
sql_exp = f"{delim_field} = 'MAJCOL' OR {delim_field} = 'MINCOL'"
# Use for loop with update cursor to gather information on new field & layer field
with arcpy.da.UpdateCursor(fc, ["COLLECTOR", "Layer"], sql_exp) as cursor: 
#using if... else to test for MAJCOL and MINCOL
    for row in cursor:
        if row[1] == "MAJCOL":
            row[0] = "MAJCOL"
        elif row[1] == "MINCOL":
            row[0] = "MINCOL"
        else:
            continue  # Skip unexpected values
# updating the row and setting collector to the correct value
        cursor.updateRow(row)

print("Update complete.")

```
